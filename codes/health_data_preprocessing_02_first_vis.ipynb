```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
pd.set_option('display.unicode.east_asian.width', True)
```

```python
# 폰트
font_path = '../malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)
```

```python
male = pd.read_excel('./datasets/Male.xlsx', index_col=0)
```

```python
grouped = male.groupby(['고유값'])
print(grouped)
```

```python
grouped_df = {}
for key, group in grouped:
  print('고유값: ', key)
  print('number: ', len(group))
  grouped_df[key] = group
  print(group.head())
```

```python
print(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].head(20))
```

```python
del grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['고유값']
del grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['성별']
del grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['생년']
del grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['검진년도']
```

```python
plt.figure(figsize = (20, 20))
sns.pairplot(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'])
plt.show()
```

```python
fig, axes = plt.subplots(len(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].columns) + 1, 1, figsize=(10,200))

for i, a in enumerate(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].columns):
    axes[i].scatter(x= grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['검진 시 연령'],
                   y= grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'][a])
    axes[i].set_title(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].columns[i])
plt.show()
```
