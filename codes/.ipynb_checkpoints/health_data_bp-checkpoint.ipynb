```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
import numpy as np
import datetime
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import Adam
```

```python
pd.set_option('display.unicode.east_asian.width', True)
```

```python
# 폰트
font_path = '../etc/malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)
```

```python
prep_male = pd.read_excel('../datasets/prep_male.xlsx', index_col= 0)
prep_female = pd.read_excel('../datasets/prep_female.xlsx', index_col= 0)
```

```python
prep_total = pd.concat([prep_male, prep_female], axis= 'rows')
```

```python
bp = prep_total[['검진 시 연령', '수축기혈압']]
```

```python
bp['검진 시 연령'].unique()
```

```python
grouped = bp.groupby(['고유값'])
grouped_df = {}
for key, group in grouped:
  print('고유값: ', key)
  print('number: ', len(group))
  grouped_df[key] = group

  print(group.head())
```

```python
for i in range(0, 75):
    grouped_df['M{}'.format(i)].reset_index(drop= True, inplace= True)
    grouped_df['M{}'.format(i)].set_index(['검진 시 연령'], inplace= True)
    grouped_df['M{}'.format(i)].to_csv('../datasets/bp/M{}.csv'.format(i))
```

```python
for i in range(0, 31):
    grouped_df['F{}'.format(i)].reset_index(drop= True, inplace= True)
    grouped_df['F{}'.format(i)].set_index(['검진 시 연령'], inplace= True)
    grouped_df['F{}'.format(i)].to_csv('../datasets/bp/F{}.csv'.format(i))
```

```python
male = pd.read_csv('../datasets/bp/M0.csv', index_col=0)
for i in range(1, 75):
    data = pd.read_csv('../datasets/bp/M{}.csv'.format(i), index_col= 0)
    male = pd.concat([male, data], axis= 'columns', ignore_index=False)
male.to_csv('../datasets/bp/test_male.csv')
```

```python
female = pd.read_csv('../datasets/bp/F0.csv', index_col=0)
for i in range(1, 31):
    data = pd.read_csv('../datasets/bp/F{}.csv'.format(i), index_col= 0)
    female = pd.concat([female, data], axis= 'columns', ignore_index=False)
female.to_csv('../datasets/bp/test_female.csv')
```

```python
test_total = pd.concat([male, female], axis= 'columns', ignore_index=False)
test_total.to_csv('../datasets/bp/test_total.csv')
```

```python
test_total.T.to_csv('../datasets/bp/test_total_T.csv')
```

```python
plt.figure(figsize=(30,15)) 
plt.plot(test_total.index, test_total.values) 
plt.savefig('../datasets/5.png')
```
