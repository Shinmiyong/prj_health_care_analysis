```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
pd.set_option('display.unicode.east_asian.width', True)
```

```python
# 폰트
font_path = '../etc/malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)
```

```python
male = pd.read_excel('../datasets/health_data.xlsx', index_col=0)
```

```python
grouped = male.groupby(['고유값'])
print(grouped)
```

```python
grouped_df = {}
for key, group in grouped:
  print('고유값: ', key)
  print('number: ', len(group))
  grouped_df[key] = group
  print(group.head())
```

```python
print(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].head(20))
```

```python
del grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['고유값']
del grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['성별']
del grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['생년']
del grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['검진년도']
```

```python
plt.figure(figsize = (20, 20))
sns.pairplot(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'])
plt.show()
```

```python
grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].columns
```

```python
fig, axes = plt.subplots(len(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].columns), 1, figsize=(10,200))

for i, a in enumerate(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].columns):
    axes[i].plot(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C']['검진 시 연령'],
                   grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'][a])
    axes[i].set_title(grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'].columns[i])
plt.show()
```

```python
fig.savefig('../datasets/1.png')
```

```python
df = grouped_df['0279C209-BEAB-493F-87EF-2290CC32074C'][['검진 시 연령', '키','체중','허리둘레','BMI','수축기혈압','이완기혈압','요단백','헤모글로빈','공복혈당','중성지방']]
```

```python
df.columns
```

```python
fig, axes = plt.subplots(len(df.columns), 1, figsize=(10,100))

for i, a in enumerate(df.columns):
    axes[i].plot(df['검진 시 연령'],
                   df[a])
    axes[i].set_title(df.columns[i])
plt.show()
```

```python
fig.savefig('../datasets/2.png')
```

```python
df.columns
```

```python
fig = plt.figure(figsize = (50, 10))
ax1 = fig.add_subplot(2,5,1)
ax2 = fig.add_subplot(2,5,2) 
ax3 = fig.add_subplot(2,5,3)
ax4 = fig.add_subplot(2,5,4)
ax5 = fig.add_subplot(2,5,5)
ax6 = fig.add_subplot(2,5,6) 
ax7 = fig.add_subplot(2,5,7)
ax8 = fig.add_subplot(2,5,8)
ax9 = fig.add_subplot(2,5,9)
ax10 = fig.add_subplot(2,5,10)

sns.pointplot(df['검진 시 연령'], df['키'], data= df, ax= ax1)
sns.pointplot(df['검진 시 연령'], df['체중'], data= df, ax= ax2)
sns.pointplot(df['검진 시 연령'], df['허리둘레'], data= df, ax= ax3)
sns.pointplot(df['검진 시 연령'], df['BMI'], data= df, ax= ax4)
sns.pointplot(df['검진 시 연령'], df['수축기혈압'], data= df, ax= ax5)
sns.pointplot(df['검진 시 연령'], df['이완기혈압'], data= df, ax= ax6)
sns.pointplot(df['검진 시 연령'], df['요단백'], data= df, ax= ax7)
sns.pointplot(df['검진 시 연령'], df['헤모글로빈'], data= df, ax= ax8)
sns.pointplot(df['검진 시 연령'], df['공복혈당'], data= df, ax= ax9)
sns.pointplot(df['검진 시 연령'], df['중성지방'], data= df, ax= ax10)

ax1.set_title('연령-키', size= 20)
ax2.set_title('연령-체중', size= 20)
ax3.set_title('연령-허리둘레', size= 20)
ax4.set_title('연령-BMI', size= 20)
ax5.set_title('연령-수축기혈압', size= 20)
ax6.set_title('연령-이완기혈압', size= 20)
ax7.set_title('연령-요단백', size= 20)
ax8.set_title('연령-헤모글로빈', size= 20)
ax9.set_title('연령-공복혈당', size= 20)
ax10.set_title('연령-중성지방', size= 20)

plt.show()
```
