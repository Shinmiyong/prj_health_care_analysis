# 남자 고유값 바꾸기

```python
male = pd.read_excel('../datasets/Male.xlsx', index_col=0)
```

```python
male.set_index(['고유값'], inplace= True)
```

```python
grouped = male.groupby(['고유값'])
grouped_df = {}
for i, (key, group) in enumerate(grouped):
  print('고유값: ', i)
  print('number: ', len(group))
  grouped_df[i] = group

  print(group.head())
```

```python
len(grouped_df)
```

```python
for i in range(0, len(grouped_df)):
    grouped_df[i].reset_index(drop=True, inplace= True)
    grouped_df[i]['고유값']= 'M' + str(i)
    grouped_df[i].set_index(['고유값'], inplace= True)
```

```python
prep_male = pd.DataFrame(grouped_df[0])
for i in range(1, len(grouped_df)):
    male_i = pd.DataFrame(grouped_df[i])
    prep_male = pd.concat([prep_male, male_i], axis = 'rows', ignore_index= False)
```

```python
prep_male.to_excel('../datasets/prep_male.xlsx')
```
